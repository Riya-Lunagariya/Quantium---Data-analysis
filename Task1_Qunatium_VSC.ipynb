```python
df = df[df["PROD_QTY"]<100]
df = df[df["TOT_SALES"]<500]
df = df.reset_index(drop=True)
df    #   Resetting the index of the pandas.DataFrame.
```

```python
from datetime import date, timedelta
start = date(1899, 12, 30)
new_date_format = []
for date in df["DATE"]:
  delta = timedelta(date)
  new_date_format.append(start + delta)

df["DATE"] = pd.to_datetime(pd.Series(new_date_format))
print(df["DATE"].dtype)
```

```python
Quantium_merged_data = pd.merge(df, purchase, on="LYLTY_CARD_NBR", how='right')   #   Merging the two pandas.DataFrames 
Quantium_merged_data
```

```python
unique_chips = Quantium_merged_data["PROD_NAME"].unique()
print(len(unique_chips))
```

```python
Quantium_merged_data = Quantium_merged_data.sort_values(by="DATE")  #   Sorting the pandas.DataFrame in ascending order of the DATE column.
Quantium_merged_data = Quantium_merged_data.reset_index(drop=True)  #   Resetting the index of the pandas.DataFrame.
Quantium_merged_data
```

```python
sales = Quantium_merged_data.groupby("DATE")["TOT_SALES"].sum().reset_index()   #Grouping the pandas.DataFrame by the TOT_SALES column for each recorded date.
sales 
```

```python
plt.figure(figsize=(10, 5))
plt.plot(sales["DATE"], sales["TOT_SALES"], color="green")
plt.title("Total Sales for the Year")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

```python
holiday_sales = Quantium_merged_data[(Quantium_merged_data["DATE"]>="2018-12-21") & (Quantium_merged_data["DATE"]<="2018-12-24")]  #   Extracting entries for the dates between 21st December 2018 and 24th December 2018.
holiday_sales=holiday_sales.sort_values(by="TOT_SALES")
holiday_sales=holiday_sales.reset_index(drop=True)  
holiday_sales
```
